```python
from medpy.io import load
image_data, image_header = load('Data/ABCs_training_data/001_t1.mha')
label_data, label_header = load('Data/ABCs_training_data/001_labelmap_task1.mha')
```

```python
image_header.get_direction()
```

```python
image_data.shape, label_data.shape
```

```python
import numpy as np
print(np.unique(label_data))
print(len(np.unique(image_data)))
np.min(image_data), np.max(image_data)
```

```python
import matplotlib.pyplot as plt

index_list = [50, 75, 90, 100, 120]
fig, ax = plt.subplots(nrows=3, ncols=len(index_list), figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')

for k, l in enumerate(index_list):
    # Show an image in each subplot
    ax[0][k].imshow(image_data[l,:,:], cmap='gist_gray')
    ax[0][k].imshow(label_data[l,:,:], cmap='jet', alpha=0.3)
    
    ax[1][k].imshow(image_data[:,l,:], cmap='gist_gray')
    ax[1][k].imshow(label_data[:,l,:], cmap='jet', alpha=0.3)

    ax[2][k].imshow(image_data[:,:,l], cmap='gist_gray')
    ax[2][k].imshow(label_data[:,:,l], cmap='jet', alpha=0.3)
```
